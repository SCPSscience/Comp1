```python
# Set variables for scatter plot
x = LASERCATS.Group
y = LASERCATS.NumberofValence

plt.scatter(x,y)
plt.title('Looking For Patterns')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

#this sets the interval on the x-axis
plt.xticks(np.arange(min(x), max(x)+1, 1.0))

# This actually shows the plot
plt.show()
```
